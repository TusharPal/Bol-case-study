# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# KNeighborsClassifier

## F1

```python
macro_f1_scores = [
    ('Base model', 0.99, 0.06, 0.97, 0.21),
    ('With stratification', 0.99, 0.20, 0.96, 0.15),
    ('With URL function', 0.99, 0.17, 0.97, 0.15),
    ('With URL components', np.nan, np.nan, np.nan, np.nan)
]

macro_f1_scores = pd.DataFrame.from_records(macro_f1_scores, 
                                            columns=['Variant', 
                                                    'Browser', 
                                                    'Hacker', 
                                                    'Robot', 
                                                    'Special'])

macro_f1_scores.set_index('Variant', inplace=True)

print(macro_f1_scores.to_string())

macro_f1_scores = macro_f1_scores.stack().to_frame('value').reset_index()

macro_f1_scores.columns = ['Variant', 'ua_agent_class', 'macro_f1']
```

```python
plt.figure(figsize=(5, 5))

sns.barplot(macro_f1_scores, 
            x='macro_f1',
            y='Variant', 
            hue='ua_agent_class', 
            orient='h')

plt.xlabel('Macro F1 score')
plt.ylabel('Model variants')

plt.legend(bbox_to_anchor=(-0.3, 0.9), loc='center right')
plt.show()
```

## Precision

```python
precision_scores = [
    ('Base model', 0.98, 0.17, 0.96, 1.00),
    ('With stratification', 0.98, 0.22, 0.96, 1.00),
    ('With URL function', 0.98, 0.59, 0.96, 0.60),
    ('With URL components', np.nan, np.nan, np.nan, np.nan),
    ('Hyperparameter tuned', np.nan, np.nan, np.nan, np.nan)
]

precision_scores = pd.DataFrame.from_records(precision_scores, 
                                            columns=['Variant', 
                                                    'Browser', 
                                                    'Hacker', 
                                                    'Robot', 
                                                    'Special'])

precision_scores.set_index('Variant', inplace=True)

print(precision_scores.to_string())

precision_scores = precision_scores.stack().to_frame('value').reset_index()

precision_scores.columns = ['Variant', 'ua_agent_class', 'precision']
```

```python
plt.figure(figsize=(5, 5))

sns.barplot(precision_scores, 
            x='precision',
            y='Variant', 
            hue='ua_agent_class', 
            orient='h')

plt.xlabel('Precision score')
plt.ylabel('Model variants')

plt.legend(bbox_to_anchor=(-0.3, 0.9), loc='center right')
plt.show()
```

# KNeighborsClassifier vs Catboost vs XGBoost
Only till URL function

```python
f1_scores = [
    ('Hacker', 0.17, 0.06, 0.08),
    ('Special', 0.15, 0.15, 0.15),
]

f1_scores = pd.DataFrame.from_records(f1_scores, 
                                        columns=['ua_agent_class', 
                                                'KNeighborsClassifier',
                                                'Catboost', 
                                                'XGBoost'])

f1_scores.set_index('ua_agent_class', inplace=True)

print(f1_scores.to_string())

f1_scores = f1_scores.stack().to_frame('value').reset_index()

f1_scores.columns = ['ua_agent_class', 'model', 'F1']
```

```python
precision_scores = [
    ('Hacker', 0.59, 0.60, 0.62),
    ('Special', 0.6, 1.00, 1.00)
]

precision_scores = pd.DataFrame.from_records(precision_scores, 
                                            columns=['ua_agent_class', 
                                                    'KNeighborsClassifier',
                                                    'Catboost', 
                                                    'XGBoost'])

precision_scores.set_index('ua_agent_class', inplace=True)

print(precision_scores.to_string())

precision_scores = precision_scores.stack().to_frame('value').reset_index()

precision_scores.columns = ['ua_agent_class', 'model', 'precision']
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, sharex=False, sharey=True, figsize=(10, 5))

sns.barplot(f1_scores, 
            x='F1',
            y='model', 
            hue='ua_agent_class', 
            orient='h',
            ax=ax[0])

sns.barplot(precision_scores, 
            x='precision',
            y='model', 
            hue='ua_agent_class', 
            orient='h',
            ax=ax[1])

plt.show()
```
